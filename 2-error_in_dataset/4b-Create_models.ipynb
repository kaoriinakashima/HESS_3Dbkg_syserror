```python
sources = ['GC_dataset_zeta_1_muoneffTrue_edispTrue',  'GC_dataset_zeta_5_muoneffTrue_edispTrue']
for source in sources:

    HESS_fixed_models =  Models.read(f'{source}/HESS_fixed_models.yml')
    print(HESS_fixed_models['HESS J1745-290'])

    # unfreezing the spectral parameters
    HESS_fixed_models_centralspectrumfree = HESS_fixed_models.copy()
    HESS_fixed_models_centralspectrumfree['HESS J1745-290'].parameters.names
    free_par_names = ['amplitude', 'index', 'lambda_']

    for free_par_name in free_par_names:
        par = HESS_fixed_models_centralspectrumfree['HESS J1745-290'].parameters[free_par_name]
        par.frozen = False
    print(HESS_fixed_models_centralspectrumfree)
    HESS_fixed_models_centralspectrumfree.write(f'{source}/HESS_fixed_models_centralspectrumfree.yml',
                                               write_covariance = False,
                                               overwrite= True)
    
```
